```python
import pandas as pd
df=pd.read_csv('Desktop\Data set\data_movie33s.csv')
df
```

```python
#Select Important columns
import numpy as np
col=['C_1_Rating','C_2_Rating','C_3_Rating','C_4_Rating']
dfs=df[col]
np.shape(dfs)
dfs
```

```python
#Normalize the dataframe for CRITIC weights
ndf=(dfs-dfs.min())/(dfs.max()-dfs.min())
ndf
```

```python
#Calculate the Standard deviation
stde=pd.DataFrame(ndf.std())
stde
```

```python
#Calculate the correlation coefficient and information
corelation=ndf.corr()
conf=1-corelation
conf
```

```python
#Calculate the CRITIC Weights
conf['M_conf']=conf.sum(axis=1)
conf['weights']=conf['M_conf']/stde[0]
conf['Nor_weights']=conf['weights']/conf['weights'].sum()
conf
```

```python
wts=[0.222509, 0.271055, 0.192073, 0.314363]
```

```python
# Normalization and Weighted Matrix for TOPSIS
nr_df=dfs/np.sqrt(np.power(dfs,2).sum(axis=0))
wdf=nr_df*wts
wdf.head()
```

```python
# calculate the positive ideal and negative ideal soluation
positive_ideal=wdf.max()
negative_ideal=wdf.min()
```

```python
#finding Separation measures
SM_P=np.sqrt(np.power(wdf-positive_ideal,2).sum(axis=1))
SM_N=np.sqrt(np.power(wdf-negative_ideal,2).sum(axis=1))
```

```python
TOP=SM_N/(SM_N+SM_P)
TOP
```

```python
df['TOPSIS']=TOP
dft=df[['User_Id','Movie_Id','TOPSIS']]
dft
```

```python
# Filter sparse movies
min_movie_ratings = 0
filter_movies = (dft['Movie_Id'].value_counts()>min_movie_ratings)
filter_movies = filter_movies[filter_movies].index.tolist()

# Filter sparse users
min_user_ratings = 19 
filter_users = (dft['User_Id'].value_counts()>min_user_ratings)
filter_users = filter_users[filter_users].index.tolist()

# Actual filtering
df_filterd_20_TOP = dft[(dft['Movie_Id'].isin(filter_movies)) & (dft['User_Id'].isin(filter_users))]
del filter_movies, filter_users, min_movie_ratings, min_user_ratings
print('Shape User-Ratings unfiltered:\t{}'.format(dft.shape))
print('Shape User-Ratings filtered:\t{}'.format(df_filterd_20_TOP.shape))
```

```python
df_filterd_20_TOP
```

```python
Mat=df_filterd_20_TOP.pivot_table(index='User_Id', columns='Movie_Id',  values="TOPSIS")
Mat
```

```python
TOPS=Mat.fillna(0)
TOPS
```
