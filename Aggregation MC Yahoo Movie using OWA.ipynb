```python
import pandas as pd
df=pd.read_csv('Desktop\Data set\data_movie33s.csv')
df
```

```python
#Select Important columns
import numpy as np
col=['C_1_Rating','C_2_Rating','C_3_Rating','C_4_Rating']
dfs=df[col]
np.shape(dfs)
dfs
```

```python
#Normalize the dataframe
N_df=(dfs-dfs.min())/(dfs.max()-dfs.min())
N_df
```

```python
#Sort the values of dataframe (in reverse order)
N_df.values.sort()
N_df
```

```python
# Multiply the ordered values with weights
# x is for a=0, b=0.5, y is for a=0.3, b=0.8, z is for a=0.5, b=1

x =[0.0, 0.0, 0.5, 0.5]
y=[0.1, 0.5, 0.4, 0.0]
z=[0.5, 0.5, 0.0, 0.0] 
x_df=x*N_df
y_df=y*N_df
z_df=z*N_df
```

```python
x_df.sum(axis=1)
y_df.sum(axis=1)
z_df.sum(axis=1)
```

```python
#Add the OWA values to the dataframe
df['x_owa']=x_df.sum(axis=1)
df['y_owa']=y_df.sum(axis=1)
df['z_owa']=z_df.sum(axis=1)
df
```

```python
xdf=df[['User_Id','Movie_Id','x_owa']]
xdf
```

```python
ydf=df[['User_Id','Movie_Id','y_owa']]
ydf
```

```python
zdf=df[['User_Id','Movie_Id','z_owa']]
zdf
```

```python
# Filter sparse movies
min_movie_ratings = 0
filter_movies = (ydf['Movie_Id'].value_counts()>min_movie_ratings)
filter_movies = filter_movies[filter_movies].index.tolist()

# Filter sparse users
min_user_ratings = 19 
filter_users = (ydf['User_Id'].value_counts()>min_user_ratings)
filter_users = filter_users[filter_users].index.tolist()

# Actual filtering
df_filterd_20_X = ydf[(ydf['Movie_Id'].isin(filter_movies)) & (ydf['User_Id'].isin(filter_users))]
del filter_movies, filter_users, min_movie_ratings, min_user_ratings
print('Shape User-Ratings unfiltered:\t{}'.format(ydf.shape))
print('Shape User-Ratings filtered:\t{}'.format(df_filterd_20_X.shape))
```

```python
df_filterd_20_X
```

```python
Mat=df_filterd_20_X.pivot_table(index='User_Id', columns='Movie_Id',  values="y_owa")
```

```python
yMat=Mat.fillna(0)
yMat
```
